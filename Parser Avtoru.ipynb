Теперь напишем функцию для сбора ссылок всех найденных объявлений

```python
# пример ссылки на объявления
# https://auto.ru/moskva/cars/skoda/all/?page=2
```

```python
'''Фунция обработчик страницы
возвращает список ссылок на объявления автомобилей Шкода с текущей страницы
page_number : int/string номер страницы для парсинга'''
def getPageLinks(page_number):
    #составляем ссылку на страницу поиска
    page_link = 'https://auto.ru/moskva/cars/used/?page={}'.format(page_number)
    
    # запрашиваем данные по ней
    response = get(page_link, headers={'User-Agent' : UserAgent().chrome})
    
    if not response.ok:
        # если сервер нам отказал, вернем пустой лист для текущей страницы
        return []
    
    # получим содержимое страницы
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    
    # ищем ссылки на наши авто объявления
    ad_links = soup.find_all('a', {'class' : 'Link ListingItemTitle__link'})
    ad_links = [item.get('href') for item in ad_links]
    
    # возвращаем все ссылки
    return ad_links
    
```

##### Напишем функцию которая будет доставать из каждой страницы необходимую информацию 

```python
def getCar(page):
    '''Возвращает словарь необходимых значений по каждому объявлению
    soup: объект bs4
    представление текущей страницы
    
    result: list
    name/price/crash/year/run/back/color/engine/tax/transmission/
    drive/rudder/condition/owners/pasport/customs/cr_date/views/comment'''
    
    # получаем html страницу в bs4
    response = get(page, headers = {'User-Agent': UserAgent().chrome})

    # проверяем 
    if not response.ok:
        # если сервер отказал вернем статус ошибки
        return response.status_code
    
    # загружаем контент
    html = response.content
    # разворачиваем дерево
    soup = BeautifulSoup(html, 'html.parser')
    
   
    # добавим имя
    try:
        name = soup.find('h1', {'class' : 'CardHead__title'})
        name = name.text
    except:
        name = None
        
    # добавим цену
    try:
        price = soup.find('span', {'class' : 'OfferPriceCaption__price'})
        price.text
        x = price.text.replace('\xa0', '')
        price = int(x.replace('₽', ''))
    except:
        price = None
    
    # добавим ДТП
    try:
        crash = soup.find('div', {'class' : 'CardBenefits__item-title'})
        crash = crash.text.replace('\xa0', ' ')
    except:
        crash = None
    
    # добавим год выпуска
    try:
        year = soup.find('a', {'class' : 'Link Link_color_black'})
        year = year.text
    except:
        year = None
        
    # добавим пробег
    try:
        run_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_kmAge'})
        run = run_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        run = int(run.text.replace('\xa0', '')[:-2])
    except:
        run = None
        
    # добавим Кузов
    try:
        back_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_bodytype'})
        back = back_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        back = back.text
    except:
        back = None
    
    # добавим Цвет
    try:
        color_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_color'})
        color = color_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        color = color.text
    except:
        color = None 
    
    # Двигатель
    try:
        engine_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_engine'})
        engine = engine_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        engine = engine.text.replace('\xa0', ' ')
    except:
        engine = None  
    
    # Налог
    try:
        tax_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_transportTax'})
        tax = tax_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        tax = tax.text.replace('\xa0', ' ')
        tax = tax.replace(' ₽ / год', '')
        tax = int(tax.replace(' ', ''))
    except:
        tax = None
        
    # Налог
    try:
        transmission_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_transmission'})
        transmission = transmission_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        transmission = transmission.text
    except:
        transmission = None 
        
                
    # Привод
    try:
        drive_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_drive'})
        drive = drive_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        drive = drive.text
    except:
        drive = None
        
    # Руль
    try:
        rudder_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_wheel'})
        rudder = rudder_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        rudder = rudder.text
    except:
        rudder = None
        
    # Состояние
    try:
        condition_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_state'})
        condition = condition_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        condition = condition.text
    except:
        condition = None
    
    # Владельцы
    try:
        owners_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_ownersCount'})
        owners = owners_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        owners = int((owners.text.replace('\xa0', ' ').split())[0])
    except:
        owners = None
        
    # ПТС
    try:
        pasport_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_pts'})
        pasport = pasport_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        pasport = pasport.text
    except:
        pasport = None
        
    # Таможня
    try:
        customs_tmp = soup.find('li', {'class' : 'CardInfoRow CardInfoRow_customs'})
        customs = customs_tmp.find_all('span', {'class' : 'CardInfoRow__cell'})[1]
        customs = customs.text
    except:
        customs = None
        
    # дата размещения
    try:
        cr_date = soup.find('div', {'class' : 'CardHead__infoItem CardHead__creationDate'})
        cr_date = cr_date.text
    except:
        cr_date = None
    
    # количество просмотров
    try:
        views = soup.find('div', {'class' : 'CardHead__infoItem CardHead__views'})
        views = int(views.text.split()[0])
    except:
        views = None
        
    # комментарий продавца
    try:
        comment = soup.find('div', {'class' : 'CardDescription__text'})
        comment = comment.find('span').get_text(' ')
    except:
        comment = None
    
    result = {'name': name, 'price': price, 'crash': crash, 'year': year,
              'run': run, 'back': back, 'color': color, 'engine': engine,
              'tax': tax, 'transmission': transmission, 'drive': drive,
              'rudder': rudder, 'condition': condition, 'owners': owners,
              'pasport': pasport, 'customs': customs, 'cr_date': cr_date,
              'views': views, 'comment': comment}
    
    return result
```

## Финальный цикл

```python
from tqdm.notebook import tqdm

final_df = pd.DataFrame(columns = ['name', 'price', 'crash', 'year',
                                   'run', 'back', 'color', 'engine',
                                   'tax', 'transmission', 'drive', 'rudder',
                                   'condition', 'owners', 'pasport', 'customs',
                                   'cr_date', 'views', 'comment'])

for page_number in tqdm(range(1, 1336), desc='Pages'):
    
    # собрали ссылки с текущей страницы
    car_links = getPageLinks(page_number)
    
    for car_link in tqdm(car_links, desc='Cars', leave=False):
        
        # иногда с первого раза страничка не прогружается
        for i in range(2):
            try:
                # пытаемся собрать данные
                data_row = getCar(car_link)
                final_df = final_df.append(data_row, ignore_index=True)
            except:
                # иначе, пробуем ещё раз
                print('AHTUNG! Parsing once again:', car_link)
                continue
```
